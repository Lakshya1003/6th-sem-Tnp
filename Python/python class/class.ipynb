<h1> Stack learning</h1>

```python
class Stack:
    clg = "LNCT"
    def __init__(self,size):
    # def __init__(self,size = None): "if we have to create a dynamic stack"
        self.stack= []
        self.size = size
        self.top = -1


    def push(self,val):
# if top == size "top is predefined 0"
# overflow : size == -1
        if self.top == self.size -1: #because we have already defined -1
            print("Overflow")
            return

        else:
            self.top += 1
            self.stack.append(val)
            return


    def pop(self):

        if self.top == -1:
            print("UNDERFLOW")
            return 

        else:
            self.stack.pop(self.top)
            self.top -= 1
            return val


    def display(self):
        print(f"Stack : {self.stack}")


    def reverse(self):
        for i in range(len(self.stack) // 2):
            temp = self.stack[i]
            self.stack[i] = self.stack[len(self.stack) - i - 1]
            self.stack[len(self.stack) - i - 1] = temp

    def himang(self):
        a = self.stack[0]
        for i in self.stack :
            if i > a:
                a = i

        return a



# "EXIT" it is the keyword use to terminate the whole program

st = Stack(6)
st.push(12)
st.push(99)
st.push(17)
st.push(15)
st.display()
st.reverse()
st.display()
st.himang()
```
